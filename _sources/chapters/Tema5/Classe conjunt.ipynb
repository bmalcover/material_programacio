# Classe Conjunt

Anem a dissenyar una classe que ens permeti representar el concepte de conjunt. Això implica que hem de pensar amb quins atributs tendrà aquesta classe.

Per altra banda, sen's demana que la classe que hem dissenyat tengui les següents operacions:

   - Un **constructor** que construeix el conjunt buit.
   - Una operació que **afegeix** un element a un conjunt si aquest element no existeix en el conjunt.
   - Una operació que **elimina** un element d'un conjunt.
   - Una funció que ens retorna el **nombre d'elements** d'un conjunt.
   - Les operacions **intersecció** i **unió**. Aquestes reben un conjunt i retornen un nou conjunt resultat d'aquesta operació.
   - La operació **pertany** que rep un element per paràmetre i retorna *True* si l'element pertany al conjunt i *False* en cas contrari.
   - Operació **és subcojunt** que rep un conjunt i retorna *True* si el conjunt de l'objecte *self* és subconjunt del conjunt rebut per paràmetre.
   - Implementar la funció *string* de Python.
   - Fer un mètode estàtic anomenat **construeix** que rep una llista i retorna un objecte del tipus Conjunt d'aquesta llista.

```python
class Conjunt:

    def __init__(self):
        # El nostre conjunt només necessita un atribut, una llista amb els elements que hi formen part.
        self.__elements = []

    def afegir(self, e):
        # Per afegir un element dins el conjunt primer hem de comprovar que encara no hi es
        if e not in self.__elements:

            self.__elements.append(e)

    def eliminar(self, e):
        # Metode que rep un element i si esta dins el conjunt l'elimina.
        if e in self.__elements:

            self.__elements.remove(e)

    def nombre_elements(self):
        # metode que ens diu quants d'elements formen part del conjunt
        return len(self.__elements)

    def unio(self, conjunt2):
        # metode que rep dos conjunts i retorna un nou conjunt que és el resultat de la seva unio
        s = Conjunt()

        for e in self.__elements:

            s.afegir(e)

        for e in conjunt2.__elements:

            s.afegir(e)

        return s

    def interseccio(self, conjunt2):
        s = Conjunt()

        for e in self.__elements:

            if e in conjunt2.__elements:

                s.afegir(e)

        return s

    def pertany(self, element):
        return element in self.__elements

    def es_subconjunt(self, conjunt2):
    # Metode que retorna True si el subconjunt identificat amb la variable self es subconjunt del conjunt2

        essubconjunt = True

        for element in self.__elements:

            if element not in conjunt2:

                essubconjunt = False

        return essubconjunt
    
    @staticmethod
    def construeix(llista):
        
        c = Conjunt()
        
        for element in llista:
            
            if not c.pertany(element):
                
                c.afegir(element)
                
        return c

    def __str__(self):
        # El metode __str__ es el mateix a totes les classes
        cadena = ""
        
        for e in self.__elements:

            cadena += str(e) + " "

        return cadena
```
