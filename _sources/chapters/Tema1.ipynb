## Conceptes bàsics

Per poder entendre bé el que representa la programació, és interessant entendre una sèrie
de conceptes bàsics relacionats amb el món de la computació, des de la seva estructura
bàsica fins aspectes relacionats indirectament amb la programació, com pot ser el
concepte de sistema operatiu o de les eines informàtiques necessàries per poder
desenvolupar programes[^f1].

### Algorismes i Programes


Un algorisme, en termes generals, és un conjunt d'instruccions que s'han de seguir per
fer una tasca. Per exemple, hi ha algorismes per cuinar (receptes), algorismes per rentar
roba (normalment als manuals de la rentadora), algorismes per solucionar problemes
matemàtics (trobar el m.c.d. i el m.c.m.), etc.

#### Història dels algorismes

Per situar l'assignatura dins el seu marc històric cal remuntar-se a molt de temps enrera,
entre els anys 3.000 i 1.500 aC. quan pareix ser que els babilonis ja empraven mètodes
algorísmics per solucionar equacions algebraiques. Els arqueòlegs varen trobar a
Mesopotàmia (Iraq) devora l'antiga ciutat de Babilònia unes taules d'argila que tenien
gravades una sèrie de passes a seguir per arribar a la solució de les equacions.
Tampoc ens podem oblidar que Euclides (300 a.c.) va proposar un algorisme pel càlcul
del màxim comú divisor de dos números enters (Elements d'Euclides, llibre 7,
proposicions 1 i 2). Es tracta de l'algorisme més antic que s'utilitza a l'actualitat.


S'ha parlat molt dels orígens de la paraula algorisme, que en un principi es va pensar
que procedia del grec _álgiros_ + _arithmos_ = algorismes ? (dolor + números). També hi va
haver qui deia que la paraula venia del Rey Algor de Castilla. Finalment els historiadors
matemàtics varen trobar l'origen autèntic de la paraula: el nom de l'autor d'un famós
llibre de text persa (_Kitab al jabr w'al-muqabala_ – regles de restauració i reducció –).
Aquest escriptor fou _Abu Ja'far Mohammed Ibn Mûsâ al-Khowârizmî_ (Pare de Jafar,
Mohamed, fill de Moisès, natural de Khowârizmî –petita ciutat de Khiva a l'antiga unió
soviètica–) que va viure a Bagdad, al segle VIII (825). Posteriorment la paraula que
originàriament era algorisme es va convertir en algoritme per “confusió apresa” amb la
paraula aritmètica.

Com es pot veure, l'estudi dels algorismes va començar al camp de les matemàtiques,
molt abans de que existissin els ordinadors. El concepte resulta molt interessant ja que
una vegada dissenyat un algorisme per una tasca, ja no cal entendre els principis en que
es bassa l'algorisme, basta amb seguir, pas per pas, les instruccions.
Una altra definició d’algorisme és aquella que el descriu com el conjunt d’operacions tal
que partint d’un estat inicial conegut permeten obtenir al cap d’un temps, finit, el
resultat esperat. Aquesta definició contempla alguns aspectes molt importants, com són
el fet de considerar l’estat d’inici, és a dir: les condicions que és necessari que es donin
per poder garantir el resultat. L’altre és el fet que es planteja que el resultat s’obté al cap
d’un cert temps, que no pot ser infinit, ja que no és raonable que s’hagi d’esperar
eternament pel resultat.

#### Programes

Quan es parla en el camp de la informàtica els algorismes es representen com a
programes. Un programa es por definir com la descripció d'un algorisme en el repertori
finit d'instruccions de la màquina. El conjunt de programes del que disposa un equip
informàtic rep el nom de programari (_software_) a diferència de l'equip informàtic en sí
que s'anomena maquinari (_hardware_). Si volem que un ordinador faci una tasca abans
hem de trobar un algorisme i després l'hem de programar a l'ordinador.
Hem de transformar l'algorisme conceptual en un conjunt clar d'instruccions. Els estudis
que se n’ocupen es basen en els nostres coneixements del llenguatge i de la gramàtica i
han generat un gran nombre d'esquemes per representar algorismes (llenguatges de
programació) fonamentats en diferents enfocaments del procés de programació
(paradigmes de programació).

### Estructura i funcionament d'un ordinador

Un ordinador és una màquina capaç d'acceptar dades a través d'un mitjà d'entrada,
processar–les automàticament sota el control d'un programa i proporcionar una
informació resultant a través d'un mitjà de sortida.

Anem a analitzar en primer lloc els components d'un ordinador des del punt de vista de
l'execució dels algorismes. Les parts fonamentals d'un ordinador són:

- **La memòria principal**, on es guarden tant els programes com les dades que
utilitzen i els resultats que generen. John von Neumann proposà una arquitectura,
una organització, dels ordinadors on a la memòria s’hi emmagatzemaven tant les
dades com les instruccions que les manipulaven aquesta arquitectura és la que
conforma la pràctica totalitat dels ordinadors actuals.
- **La Unitat Central de Procés (CPU)** que es troba connectada amb la memòria i
s'encarrega de manipular les dades i de controlar l'ordre d'execució de les
instruccions. Així, la CPU agafa instruccions de la memòria principal i les
executa, l'execució fa que la CPU agafi les dades de la memòria principal, les
processi i en cas necessari deixi els resultats a la memòria per guardar-los.
- **Els dispositius perifèrics** que s'encarreguen de l'entrada (teclat, rata, tauleta, càmera,
etc.) i la sortida (pantalla, impressora, etc.) de dades i de l'emmagatzematge
massiu (unitats de disc i cintes magnètiques).


#### La memòria
La memòria es pot considerar organitzada en memòria principal i memòria secundària.
Hi ha dos tipus de memòria principal: la ROM (memòria només de lectura on el
fabricant posa les instruccions bàsiques per fer funcionar l'ordinador) i la RAM
(memòria de lectura i escriptura, en aquesta es on s'hi pot escriure informació). Tota la
informació que es tractada per la CPU ha de trobar–se necessàriament a la memòria
principal. Pot considerar–se la memòria principal com si es tractés d'un conjunt de
caselles on a cada una d'elles s'hi troba informació. Es pot accedir a cada una de les
caselles de manera individual.
La memòria secundària, que es situa sobre els sistemes d'emmagatzematge massiu,
guarda la informació que no es necessita immediatament per poder fer feina.

#### La Unitat Central de Procés o CPU

La CPU té dues components fonamentals:
- **La unitat de control (UC)**: és la unitat que controla les operacions per fer el
tractament de la informació, interpreta les instruccions del programa emmagatzemat a la memòria i dóna les ordre
corresponents a la resta dels Components de l'ordinador.

- **La unitat aritmètica i lògica (UAL)**: és la part de la CPU que realitza les
operacions matemàtiques i lògiques amb les dades que rep de la unitat de
control. Les operacions que pot realitzar són bastant senzilles, la complexitat
de les operacions que pot realitzar un ordinador es basa en l'enorme velocitat a
la que es poden dur a terme aquestes operacions.

A la CPU també es troben els registres, que són llocs on es guarden les dades a tractar
per cada operació mentre aquesta es realitza. La informació va des de la memòria a
aquests registres per ser tractada, i després torna una altra vegada a la memòria.

#### Els bussos de comunicacions

Els components estan interconnectats amb línies de comunicacions que es coneixen com
a bussos. Hi ha el bus d'adreces pel que circulen les adreces que identifiquen de quina
posició de memòria es vol llegir informació o a quina posició de memòria es vol
escriure informació i el bus de dades pel que circulen les dades que s'han de llegir o
s'han d'escriure.


### Unitats d'informació

Els ordinadors funcionen, bàsicament, amb un sistema de transistors (que venen a ser
una espècie d'interruptors). Per tant, la informació que poden entendre els ordinadors es
redueix exclusivament al pas o no de corrent elèctric per cada un d'aquests transistors. A
aquesta unitat d'informació mínima que poden entendre els ordenadors se l'anomena bit.
Evidentment, amb un sol bit d'informació es pot dir poca cosa, per això és que
usualment s'agrupen de manera que es pugui representar més informació. Així, si amb
un bit només es pot diferenciar entre dos possibles estats (1 – pas de corrent, i 0 – no
pas de corrent), amb aquestes agrupacions es poden fer combinacions; per exemple: si
s'agrupen 8 bits aleshores es poden representar 256 possibles combinacions de bits
( $C_{2}^{8}$ = $256$ ), si s'agrupen 16 bits el nombre de combinacions serà de 65.536
( $$C_{2}^{16}$$ = $$65536$$  ). Anomenam **Byte** les agrupacions de 8 bits. Quan la informació
va i ve pels bussos, ho fa en grups de bits, la grandària d'aquests grups és la quantitat
mínima d'informació que pot tractar un ordinador determinat, aquests grups es coneixen
com paraula o paraula de memòria.
L'ordinador està preparat per executar operacions de molt baix nivell (codi màquina)
que tenen el principal inconvenient de que són difícils de comprendre i d'utilitzar. És per
això que es varen desenvolupar els anomenats llenguatges de programació d'alt nivell.
L'evolució dels llenguatges de programació, igual que es fa amb els ordinadors, es pot
descriure en diferents generacions:


### Història dels llenguatges de programació


#### Primera generació

Llenguatges de màquina o de primera generació, on els programadors havien de
realitzar els seus programes directament amb zeros i uns. S'utilitzava el sistema
hexadecimal per assimilar millor les instruccions.

#### Segona generació

Llenguatges ensambladors o de segona generació en el que les instruccions continuen
essent les mateixes, únicament canvia la sintaxi utilitzada. S'assignen noms mnemònics
als diversos codis d'operació i aquests s'utilitzen en lloc dels codis hexadecimals durant
el procés de disseny. Les instruccions dels programes s'expressen en termes d'una
màquina específica.

#### Tercera generació

Llenguatges d'alt nivell o de tercera generació on es defineixen una sèrie de primitives
d'alt nivell dissenyades com a una seqüència de les primitives de baix nivell disponibles
al llenguatge de màquina. No es fa referència a com ha de fer la tasca directament la
màquina. Els traductors d'aquests llenguatges havien de reunir (o recopilar o compilar
!!) varies instruccions de màquina per simular l'activitat demanada per una primitiva
d'alt nivell. Per això es va començar a anomenar compiladors als programes traductors.
És en aquest nivell on s'aconsegueix la independència respecte als ordinadors.
Únicament calia disposar del traductors adequats per a cada màquina. Per la falta de
consens a l'hora de definir correctament un llenguatge existeixen el American National
Standards Institute (ANSI) i la International Standards Organization (ISO).

El llenguatge `Python` que emprarem aquest curs per aprendre a programar pertany a aquesta tercera generació, va ser
creat per Guido van Rossum l'any 1991.



#### Quarta generació

Llenguatges de quarta generació on s'agrupen tots els paquets de software que
permeten als usuaris adaptar els programes a les seves aplicacions sense haver de tenir
coneixements tècnics. S'hi inclouen les fulles de càlcul, els sistemes de bases de dades,
els paquets gràfics, els processadors de texts, etc. Són aplicacions que es solen agrupar
en paquets formant sistemes que, per exemple, permeten a una persona controlar
fàcilment un petit negoci. Es considera una altra generació ja que l'entorn de
programació que ofereixen és molt més proper al de l'aplicació que els entorns oferts
pels llenguatges de tercera generació.

#### Cinquena generació

Llenguatges de cinquena generació és un terme que s'ha utilitzat molt per descriure el
concepte de programació declarativa (en concret la programació lògica). Es basa en el
fet de que un programador s'ha de concentrar únicament en la naturalesa del problema i
no en la manera de solucionar–ho. S'aprofita el sistema de càlcul de l'ordinador per què
a partir del nostre plantejament es derivi la solució cercada.

Molts dels llenguatges d'alt nivell són fonamentalment algebraics ja que aquestes
notacions són més adequades que el llenguatge natural per a descriure els processos
computacionals (o programes).

Les propietats més importants que diferencien aquests llenguatges dels llenguatges
màquina són:
- No necessiten que l'usuari conegui detalls com l'existència dels registres, les
representacions internes de les dades, els canals d'entrada/sortida, etc.
- Ofereixen la possibilitat de moure els programes d'un ordinador a l'altre. Hi ha
un cert grau d'independència de l'ordinador o del sistema.
- Permeten escriure els programes d'una manera més concisa.
- Permeten que els programes s'escriguin en termes orientats a problemes. Per
exemple, es poden donar noms simbòlics a les dades i es poden incloure
operadors matemàtics i expressions dins els llenguatges de programació.

### Interprets i Compiladors

El nivell d'abstracció d'un llenguatge d'alt nivell no es correspon, com és lògic, amb el
del llenguatge de baix nivell al que vol substituir. Com a conseqüència, no hi ha una
equivalència una a una entre les instruccions dels dos nivells. Per què l'ordinador pugui
comprendre el que el llenguatge d'alt nivell representa ens cal una traducció.
Cadascuna de les instruccions d'alt nivell ha de ser traduïda en un conjunt d'instruccions
de baix nivell que facin la mateixa tasca que la primera. Aquesta traducció es pot fer de
manera anàloga a com es traduiria un fragment d'una llengua estrangera al català. Si el
que es vol és disposar d'una traducció permanent, com ara la traducció d'un llibre que
ens agradaria llegir en la nostra llengua, es pot fer primer tot el procés complet de
traducció i després es pot executar el procés –o llegir el llibre traduït. En aquest cas
tendríem el que es coneix com a una compilació. Si per una altra banda es vol
simplement conèixer el que es diu en un moment determinat, com la traducció
simultània d'una conferència d'algú de parla estrangera que ens agradaria escoltar, es pot anar traduint i executant
–o escoltant– simultàniament. És el que es coneix com a
interpretació.



#### Sistemes Operatius

Hi ha una unitat de software molt important que és l'anomenat Sistema Operatiu. Quan s'encén l'ordinador s'executa
automàticament el Sistema Operatiu, que ens permet emprar el teclat i el ratolí per introduir ordres bàsiques
(execució d'un programa d'aplicació com un processador de texts, formatejar un disc, copiar informació d'un disc a un
altre, etc.). A més a més de la interacció amb el món exterior, el sistema operatiu també s'encarrega de gestionar
una utilització òptima dels recursos de que es disposa. Entenem per recursos coses com els dispositius perifèrics,
l'espai de la memòria principal, la CPU, etc.
Avui en dia entre els sistemes operatius que és s'utilitzen tenim el sistema Unix (sobretot pels ordinadors grans,
les estacions de treball i ara amb els dispositius mòbils), el Windows en totes les seves versions (per ordenadors
personals), el Linux (versió d'Unix per ordinadors personals) i el MacOS X per ordinadors Apple.


### Procés de fer un programa

Algunes de les eines que habitualment proporciona el sistema operatiu són les necessàries per a l'escriptura, la
traducció i l'execució dels programes. La primera passa a l'hora d'escriure un programa és l'edició del que s'anomena
programa font. És necessari disposar del que s'anomena un editor de texts per poder escriure, amb la sintaxi del
llenguatge de programació adient, el programa. Una vegada que es disposa del programa font, un programa compilador
s'encarrega de la seva traducció al llenguatge de la màquina.

### El procés de compilació
Si el procés va be i el programa és correcte lèxicament, sintàctica i semàntica, s'obté el que s'anomena programa en
codi font. En cas contrari, el compilador avisa al programador dels possibles errors que s'han trobat i el procés es
repeteix: el programa es torna a editar per tal de corregir els errors i després es torna a compilar.

El programa en codi font encara no és qualque cosa que es pugui utilitzar directament. És necessari afegir–hi totes
les coses que el compilador ha afegit al codi. Per exemple, és necessari afegir tots els mitjans per poder realitzar
una operació de lectura del teclat o escriptura a la pantalla. Això és així per què hi ha operacions que no
necessàriament es fan igual a tots els sistemes, depenen del sistema operatiu. Per tant, és necessari realitzar
encara una altra etapa, el muntatge, on un programa anomenat linker s'encarregarà d'afegir totes aquestes coses. A
aquesta fase també és possible que es produeixin errors, en tal cas el programador haurà de tornar a repetir el
procés d'editar el codi, fer els canvis adients, tornar a compilar i després tornar a muntar. Si no es produeix cap
error, finalment s'obté el programa executable. A l’actualitat tot el procés es pot desenvolupar des del que es
coneix com **Entorns Integrats de Desenvolupament** (_IDE_ per les seves sigles en anglé) que són
aplicacions des de les quals es poden realitzar totes i cada una de les passes esmentades i fins i tot es pot realitzar
una execució dels programes de manera que si es produeix qualque tipus d’error es mostra la línia de codi que l’ha
provocat.

### El procés de Python

En aquest curs el llenguatge de programació que usarem és `Python`, que és un llenguatge dissenyat per què les
aplicacions que s’escriguin amb ell siguin completament independents de la plataforma. Això significa que un programa
 escrit en `Python` s’executarà sense cap tipus de modificació en absolutament qualsevol sistema informàtic, al menys per
qualsevol sistema que suporti `Python`.


En la seva major part, `Python` és un llenguatge interpretat i no un compilat, tot i que la compilació és un pas. El
codi `Python`, escrit en un fitxer .py, es compila primer amb el que s’anomena bytecode (que es descriurà amb més detall)
que s’emmagatzema amb un format .pyc.

`Python` en primer lloc  compila el codi font escrit en un fitxer _.py_  en un format conegut com a _byte code_
emmagatzemat en fitxer amb format _.pyc_. La compilació en aquest cas és simplement el pas de realitzar la traducció
del codi font del llenguatge d'alt nivell (el que aprendrem durant el curs) a una representació de nivell inferior
independent de la plataforma.

El _byte code_ (fitxer .pyc) es carrega en el moment que volem executar un programa de `Python` i és interpretat per una
màquina virtual de `Python` (PVM), que no és mes que un programa que llegeix cada instrucció del _byte code_ i
executa l'operació que toca a l'ordinador.

Sempre que realitzam una instal·lació de l'entorn de ´Python´ el PVM s'instal·la com una part seva, ja que és el component
que realment executa els nostres programes. Tècnicament aquest és només l’últim pas del que s’anomena l'intèrpret de
'Python'. A més , durant el procés es duen a terme optimitzacions que queden fora del nostre abast per millorar el
rendiment del programa.

Cada vegada que s’executa un programa interpretat, l’intèrpret ha de convertir el codi font en codi de màquina, tal
com hem descrit anteriorment, però també s'encarrega  d’introduir altres components necessàris per l’execució del
programa, com poden ser lliberires de codi adiccional.


#### Inconvenients dels llenguatges interpretats

Aquest procés de conversió fa que un programa interpretat funcioni més lent que un programa similar
escrit en un llenguatge compilat. Python fa alguna cosa intel·ligent per millorar el seu rendiment, el fet de
compilar a bytecode (fitxers .pyc) la primera vegada que executa un programa,  millora substancialment les futures
execucions del mateix programa.

[^f1]: Aquest capítol està escrit per en Ramon Mas i en Pere Palmer en un material encara no
publicat
