## Exercicis Tema 2


Aquests exercicis s'han de realitzar mitjançant aquests 3 passos:

- **Pensar**: Abans de programar cal tenir clar quina solució aplicarem, intentar relacionar el problema amb aquells
continguts que hem vist a classe.
- **Programar**: Aquesta és la part que per ens portarà més temps, ja que encara no coneixem tot el lèxic del
llenguatge
ni les seves característiques. En aquestes primeres setmanes és normal programar de forma lenta i amb molts errors.
- **Avaluar**: Un cop hem acabat de realitzar el programa, hem de poder assegurar que el codi funciona tal com esperem,
per a això realitzarem proves amb diferents valors d'entrada per assegurar-nos que el codi funciona.

### Exercicis bàsics

En aquests exercicis no es realitza lectura de teclat, els valors s'assignen dins del programa

1.Mostra per pantalla el valor que tendrà la variable x després d’haver-li assignat l’expressió 5 + 4.

2.Definir dues variables anomenades x i y a les quals els assignarem un valor enter i un valor decimal
respectivament. El programa ha de mostrar el resultat de sumar, restar, multiplicar i dividir-los (divisió sencera i decimal).

3.Creau dues variables booleanes anomenades b1 i b2, assignau el valor True a b1 i el valor False a b2. Mostrau el
resultat de les següents expressions per pantalla:

```
    b1 and b2
    b1 or b2
    (b1 and b2) or (b1 or b2)
    not b1
```

4.Escriure una expressió que creï una variable, representant una temperatura en graus Celsius i la transformi a
graus Fahrenheit. Escriu una segona expressió que transformi metres en milles.

### Bàsics amb entrada/sortida

Introduïm sentències de control: if/ else. Aquests exercicis fan entrada per teclat i mostren la sortida per pantalla.


5.Implementar un programa que llegeix dos nombres enters i mostra el quocient de la divisió i la resta.

6.Implementar un programa que sigui capaç de resoldre una equació de primer grau de la forma `ax+b= 0`. Ha de llegir 2
nombres del teclat corresponents a `a` i `b` i mostrar la solució per pantalla.

7.Implementar un programa que llegeix un nombre real mostra si el aquest és positiu.

8.Implementar un programa que llegeix un nombre real i mostra si el nombre està en el rang (-1.0, 1.0).

9.Escriu un programa que llegeixi les coordenades d'un punt (2 valors) i indiqui en el quadrant (del sistema
cartesià) en el qual es troba aquest punt. Consideram el primer quadrant aquell on els dos valors són positius i
rotarem com les manetes del rellotge, si una de les components és 0 el programa mostrarà la frase "Quadrant
indeterminat".

10.Escriu un programa que rebi tres valors d'entrada que corresponen a hores, minuts i segons. El programa ha
d'indicar si es tracta d'un rellotge amb una hora vàlida.

```
    Si rebem els nombres 9, 56, 33 és correcte.
    Si rebem els nombres 9, 64, 22 no és correcte ja que no existeixen els 64 minuts.
    Si rebem els nombres 25, 44, 22 no és correcte ja que no existeix l'hora 25.
```
11.Escriu una expressió que ens permeti determinar si un valor de tipus enter positiu pot correspondre a un any de
traspàs. Es considera any de traspàs si el nombre és divisible per quatre excepte els anys que són múltiples de cent
llevat que ho siguin de 400.

12.Escriure un programa per comprovar si tres valors decimals es corresponen amb un triangle, a més ha de dir si
aquest és equilàter, isòsceles o escalè.

13.Implementar un programa que llegeixi la nota d'un alumne, és a dir, un valor entre 0.0 i 10.0, i tot seguit
mostri la qualificació: Suspès, Aprovat, Notable, Excel·lent, Matrícula d'Honor. Heu d'utilitzar els següents
intervals 0 - 5.0, 5.0 - 7.0, 7.0 - 8.5, 8.5 - 9.5, 9.5 - 10.0

14.Escriu un programa per calcular l'edat d'un gos en anys canins. Durant els dos primers anys, un any caní és igual
a 10.5 anys humans. Després d'això, cada any caní equival a 4 anys humans.  _Exemple_: 12 anys humans corresponen a 61
anys canins.

15.Implementar un programa que determini si un nombre sencer és múltiple de 2.

16.Implementar un programa que demani dos nombres enters i determini si el primer nombre és múltiple del segon.

17.Implementar un programa que demani l'any de naixement a l'usuari i li mostri el seu horòscop xinès. L'usuari ha
de poder introduir qualsevol any.
```
    Rata:   1924, 1936, 1948....
    Buey:   1925, 1937, 1949....
    Tigre:  1926, 1938, 1950....
    Conejo: 1927, 1939, 1951....
    Dragón: 1928, 1940, 1952....
    Serpiente: 1929, 1941, 1953....
    Caballo: 1930, 1942, 1954....
    Cabra:  1931, 1943, 1955....
    Mono:   1932, 1944, 1956....
    Gallo:  1933, 1945, 1957....
    Perro:  1934, 1946, 1958....
    Cerdo:  1935, 1947, 1959....
```

L'usuari ha de poder introduir qualsevol any. Per exemple, si l’usuari introdueix l’any 15895, i el programa ha
d'indicar Conill; si introdueix 1024 ha de mostrar Rata.


### Bucles

Important: Abans d'implementar els exercicis esbossar la solució sobre paper.
Introduïm sentències de control: MENTRE i PER

18.Escriu un programa per esbrinar si un nombre natural m és un quadrat perfecte, és a dir, si és de la forma $m = n^2$
per algún nombre natural $n$.

19.Escriure un programa, que llegit un nombre per teclat (1 a 9), et mostri per pantalla la seva taula. Per exemple:

```
    2←┘
    La taula del 2 és:
    2 x  0 = 0
    2 x  1 = 2
    2 x  2 = 4
    2 x  3 = 6
    2 x  4 = 8
    2 x  5 = 10
    2 x  6 = 12
    2 x  7 = 14
    2 x  8 = 16
    2 x  9 = 18
    2 x 10 = 20
```

20.Implementar un programa on l'usuari va introduint nombres enters. El programa ha d'anar imprimint per pantalla si
 el nombre és parell o imparell per a cada valor introduït. Si introdueix el valor 0, el programa ha de finalitzar
 sense indicar si el nombre és parell / senar.

21.Millorar el programa anterior perquè a l'acabar, imprimeixi per pantalla el nombre de vegades que l’usuari ha
posat nombres parells i el nombre de vegades que ha posat números senars. p.e:
```
    2←┘ parell
    3←┘ senar
    8←┘ parell
    122←┘ senar
    0←┘
    Total de nombres parells: 3 Total de nombres senars: 1
```

22.Implementar un programa que calculi tots els divisors d'un nombre. Per exemple: els divisors de 18 són; 18, 9, 6,
3, 2 i 1

23.Escriu un programa que calculi la mitjana d'un conjunt de nombres. El programa demanarà a l'usuari quants nombres
 es faran servir. L'usuari ha d'introduir aquesta quantitat de nombres, el programa en calcularà i mostrarà la mitjana.

24.Escriu un programa que llegeix un nombre del teclat i que indiqui si és un nombre primer o no.

25.Escriure un programa que mostri per pantalla tots els nombres que són divisibles per 7, però no són múltiple de 5
i es troben en el rang(2000, 3000).

26.Donat un nombre enter N, si la suma dels seus divisors (sense comptar el propi nombre) és igual a `n` es diu que
aquest nombre és perfecte. Si aquesta suma és inferior, es diu que és deficient. Si és superior es diu que és
abundant.
```
    6 té com a divisors: 1, 2, 3. La seva suma és 6, per tant és perfecte.
    8 té com a divisors: 1, 2, 4. La seva suma és 7, pel que és deficient
    24 té com a divisors:1, 2, 3, 4, 6, 8, 12. La seva suma és 36, per tant és abundant.
```

### Més pràctica

27. Equació de segon grau Ax2 + Bx + C = 0 i calculi les seves dues solucions. Se suposa que l'equació té solucions
reals. Òbviament, les dades del programa són els coeficients de l'equació: A, B i C.
Per trobar l'arrel quadrada d'un nombre es fa amb la funció _sqrt_:

```
    import math
    #S'ha de posar aquesta línia, abans de començar el codi.

    f = math.sqrt(x) # on x es una variable sencera.
```

programa b:

    2. Si és imparell, se li multiplica tres i se li suma un;
    3. La successió acaba en arribar a 1.

Feu un programa que demani un sencer a l'usuari i calculi la seva seqüència de Collatz.
