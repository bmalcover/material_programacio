## Exercicis Tema 2


Aquests exercicis s'han de realitzar mitjançant aquests 3 passos:

- **Pensar**: Abans de programar cal tenir clar quina solució aplicarem, intentar relacionar el problema amb aquells
continguts que hem vist a classe.
- **Programar**: Aquesta és la part que ens portarà més temps, ja que encara no coneixem tot el lèxic del
llenguatge ni les seves característiques. En aquestes primeres setmanes és normal programar de forma lenta i amb
molts errors.
- **Avaluar**: Un cop hem acabat de realitzar el programa, hem de poder assegurar que el codi funciona tal com esperem,
per a això realitzarem proves amb diferents valors d'entrada per assegurar-nos que el codi funciona.

### Exercicis bàsics

En aquests exercicis no es realitza lectura de teclat, els valors s'assignen dins del programa:

1-Mostrar per pantalla el valor que tindrà la variable x després d’haver-li assignat l’expressió 5 + 4.

2-Definim dues variables anomenades `x` i `y` a les quals els assignarem un valor enter i un valor en coma flotant
respectivament. El programa ha de mostrar el resultat de sumar, restar, multiplicar i dividir-les (divisió sencera i
decimal).

3-Crear dues variables booleanes anomenades `b1` i `b2`, assignar el valor `True` a `b1` i el valor `False` a `b2`.
Es demana que mostreu el resultat de les següents expressions per pantalla:

```
    b1 and b2
    b1 or b2
    (b1 and b2) or (b1 or b2)
    not b1
```

4-Escriure un programa que contengui el següent: una expressió que creï una variable representant una temperatura en
graus Celsius i una segona expressió que la transformi a graus Fahrenheit. Escriu un segon programa que realitzi una
feina similar, transformant metres en milles.

### Bàsics amb entrada/sortida

Introduïm les sentències de control: if/ else. Aquests exercicis fan entrada per teclat i mostren la sortida per
pantalla.

5-Implementar un programa que llegeix dos nombres enters i mostra el quocient de la divisió i la resta.

6-Implementar un programa que sigui capaç de resoldre una equació de primer grau de la forma `ax+b= 0`. Ha de llegir 2
nombres del teclat corresponents a `a` i `b` i mostrar la solució per pantalla.

7-Implementar un programa que llegeix un nombre real mostra per pantalla un missatge que indica si aquest és positiu.

8-Implementar un programa que llegeix un nombre real i mostra per pantalla un missatge que indica si el nombre està en
 el rang (-1.0, 1.0) o no.

9-Escriu un programa que llegeix les coordenades d'un punt en 2D i indiqui en el quadrant (del sistema
cartesià) en el qual es troba aquest punt. Consideram el primer quadrant aquell on els dos valors són positius i
girarem com ho fan les manetes del rellotge. Si una de les components és 0, el programa mostrarà la frase: "Quadrant
indeterminat".

10-Escriu un programa que rebi tres valors d'entrada que corresponen a hores, minuts i segons. El programa ha
d'indicar si es tracta d'un rellotge amb una hora vàlida.

```
    Si rebem els nombres 9, 56, 33 és correcte.
    Si rebem els nombres 9, 64, 22 no és correcte ja que no existeixen els 64 minuts.
    Si rebem els nombres 25, 44, 22 no és correcte ja que no existeix l'hora 25.
```
11-Escriu una expressió que ens permeti determinar si un valor de tipus enter positiu pot correspondre a un any de
traspàs. Es considera any de traspàs aquell nombre que és divisible per quatre excepte els anys que són múltiples de
cent, llevat que ho siguin de 400.

12-Construeix un programa per comprovar si tres valors decimals es corresponen amb un triangle, això vol dir que el
programa ha de comprovar si la suma de dos dels seus costats és més gran a la longitud del costat restant. En cas
positiu, el programa ha de dir si aquest triangle és equilàter, isòsceles o escalè.

13-Implementa un programa que llegeixi la nota d'un alumne, és a dir, un valor entre 0.0 i 10.0, i tot seguit
mostra la seva qualificació: Suspès, Aprovat, Notable, Excel·lent o Matrícula d'Honor. Heu d'utilitzar els següents
intervals: 0 - 5.0, 5.0 - 7.0, 7.0 - 8.5, 8.5 - 9.5, 9.5 - 10.0.

14-Escriu un programa per calcular l'edat d'un gos en anys canins. Durant els dos primers anys, un any caní és igual
a 10.5 anys humans. A partir del tercer, cada any caní equival a quatre anys humans. _Exemple_: 12 anys humans
corresponen a 61 anys canins.

15-Implementar un programa que determini si un nombre sencer és múltiple de 2.

16-Implementar un programa que demani dos nombres enters i determini si el primer nombre és múltiple del segon.

17-Dissenyar un programa que demani l'any de naixement a l'usuari i li mostri el seu horòscop xinès. L'usuari ha
de poder introduir qualsevol any:
```
    Rata:   1924, 1936, 1948....
    Bou:   1925, 1937, 1949....
    Tigre:  1926, 1938, 1950....
    Conill: 1927, 1939, 1951....
    Drac: 1928, 1940, 1952....
    Serp: 1929, 1941, 1953....
    Cavall: 1930, 1942, 1954....
    Cabra:  1931, 1943, 1955....
    Mico:   1932, 1944, 1956....
    Gall:  1933, 1945, 1957....
    Ca:  1934, 1946, 1958....
    Porc:  1935, 1947, 1959....
```

L'usuari ha de poder introduir qualsevol any. Per exemple, si l’usuari introdueix l’any 15895, i el programa ha
d'indicar Conill; si introdueix 1024 ha de mostrar Rata.


### Bucles

Important: Abans d'implementar els exercicis esbossar la solució sobre paper.
Introduïm sentències de control: MENTRE i PER

18-Escriu un programa per esbrinar si un nombre natural `m` és un quadrat perfecte, és a dir, si és de la forma $m =
n^2$ per algun nombre natural $n$.

19-Escriure un programa que sigui capaç de llegir un nombre d'un sol dígit per teclat i mostri per pantalla  la seva taula
de multiplicar.
Per exemple:
```
    2←┘
    La taula del 2 és:
    2 x  0 = 0
    2 x  1 = 2
    2 x  2 = 4
    2 x  3 = 6
    2 x  4 = 8
    2 x  5 = 10
    2 x  6 = 12
    2 x  7 = 14
    2 x  8 = 16
    2 x  9 = 18
    2 x 10 = 20
```

20-Implementar un programa on l'usuari introdueix una seqüència de nombres enters. El programa ha d'anar imprimint
per pantalla si el nombre és parell o senar per a cada valor introduït. Si introdueix el valor 0, el programa ha
de finalitzar sense indicar si el nombre és parell / senar.

21-Millorar el programa anterior perquè a l'acabar, imprimeixi per pantalla el nombre de vegades que l’usuari ha
posat nombres parells i el nombre de vegades que ha posat nombres senars. Per exemple:
```
    2←┘ parell
    3←┘ senar
    8←┘ parell
    122←┘ senar
    0←┘
    Total de nombres parells: 3
    Total de nombres senars: 1
```

22-Implementar un programa que calculi tots els divisors d'un nombre. Per exemple: els divisors de 18 són; 18, 9, 6,
3, 2 i 1-

23-Escriu un programa que calculi la mitjana d'un conjunt de nombres. En primer lloc, el programa demanarà a l'usuari
quants nombres es faran servir. Després, l'usuari ha d'introduir aquesta quantitat de nombres i el programa en
calcularà i mostrarà la mitjana per pantalla.

24-Escriu un programa que llegeix un nombre del teclat i que indiqui si és un nombre primer o no.

25-Escriure un programa que mostri per pantalla tots els nombres que són divisibles per 7, però no són múltiples de 5
i es troben en el rang: (2000, 3000).

26-Donat un nombre enter `n`. Si la suma dels seus divisors (sense comptar el mateix nombre) és igual a `n`, es diu que
aquest nombre és perfecte. Si aquesta suma és inferior, es diu que és deficient. Si és superior es diu que és
abundant. Realitza un programa que donat un nombre enter ens digui si és perfecte, deficient o abundant.
```
    6 té com a divisors: 1, 2, 3, la seva suma és 6, per tant és perfecte.
    8 té com a divisors: 1, 2, 4, la seva suma és 7, pel que és deficient
    24 té com a divisors:1, 2, 3, 4, 6, 8, 12, la seva suma és 36, per tant és abundant.
```

### Més pràctica

27-Volem fer un programa que té la capacitat de resoldre equacins de segon grau (Ax2 + Bx + C = 0) i calculi les seves
dues solucions. Se suposa que l'equació té solucions reals. Òbviament, les dades que ha de rebre el programa són els
coeficients de l'equació: `A`, `B` i `C`. Per trobar l'arrel quadrada d'un nombre es fa amb la funció _sqrt_ tal com
podem observar a continuació:
```
    # Abans de començar el codi, s'ha de posar aquesta línia
    import math

    f = math.sqrt(x) # on x es una variable de tipus nombre enter.
```

28- Anem a dibuixar asteriscs amb els següents exercicis:

a.	Escriu un programa que demani a l'usuari ingressar l'altura i l'amplada d'un rectangle i el dibuixi utilitzant
asteriscs.

b.	Escriu un programa que dibuixi el triangle de la mida indicat per l'usuari. A continuació teniu un exemple del
programa b:

```
    Alçada: 5
    *
    **
    ***
    ****
    *****
```
29-La seqüència de Collatz d'un nombre enter es construeix de la següent manera:

    1. Si el nombre és parell, es divideix per dos.
    2. Si és imparell, se li multiplica tres i se li suma un;
    3. La successió acaba en arribar a 1.

Feu un programa que demani un nombre enter a l'usuari i calculi la seva seqüència de Collatz.
