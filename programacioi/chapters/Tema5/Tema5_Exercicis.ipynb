#### Strings

Aquests exercicis están dissenyats per practicar només amb _Strings_. Podeu emprar tots els mètodes existents.

1.Donat un _String_ llegit per teclat, crea un nou _String_ on s'hagin eliminat tots els caràcters excepte els nombres enters. 
Exemple: `str1 = "Tinc 25 anys i 10 mesos"` llavors `str = "2510`.

2.Donats dos _Strings_, `s1` i `s2`. Escriu un programa per crear un nou _String_ anomenat s3 que és el resultat d'inserir `s2` al mig de _s1_. Si la longitut de s1 és senar, la part més llarga quedarà al principi.  Exemple: `s1 = "joan"` i `s2=patata` llavors `s3=jopatataan`.

3.Escriu una funció anomenada `digits` que rep un _String_, `s1`, que retorna la suma dels dígits que apareixen a `s1`, ignorant tots els altres caràcters. Recorda que en aquests exercicis només pots emprar mètodes relacionats amb _Strings_.

4.Escriu un programa per trobar paraules que contenen lletres de l'alfabet i números (ambdues coses) a partir d'un _String_ d'entrada.

Exemple: `str1 = "Emma25 is 50scientist and AI Expert"`  llavors ha de mostrar `Emma25` i `50scientist`.

5.Crear un programa que llegeix una seqüència de paraules que acaben en la paraula buida; Determinar si cada paraula és un palíndrom (capicua).

#### Diccionaris

1.Fer un programa que llegeix un _String_ de teclat i mostra la freqüència d'aparició de cada lletra del diccionari. El nostre programa no sap diferenciar majúscules de minúscules.

2.Donat el diccionari resultant del programa anterior crea un nou diccionari on les claus siguin les freqüències d'aparició i els valors siguin llistes de lletres.

3.Escriure una funció anomenada `compta_paraules` que rep un _String_ amb paraules separades per espais en blanc i retorni un diccionari amb el recompte de cada paraula. 

4.Escriu una funció anomenada `divisors` que rep una llista d'enters positius i retorni un diccionari que associa a cada nombre la seva llista de divisors.

5.Fes una funció anomenada `matriu2diccionari` que rep una matriu per paràmetre i retorna un diccionari on les claus són tuples `(i, j)` i els valors els nombres diferents de zero. Escriu una segona funció que donat un diccionari i dos indexos `i` i `j` permeti obtenir el valor en una posició (i, j).

6.Donat un diccionari que representa una funció (les claus són els elements del domini, i els valors del codomini), escriu una funció que determini si aquesta funció és injectiva. Una funció és injectiva si no hi ha dos valors de les claus que s’assignin al mateix valor.
